#Market Basket Analysis(ASSOCIATION RULE)

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mlxtend
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data = pd.read_csv('/content/basket_analysis.csv')
data.head()
```

```python
#dropping the unnamed column
data.drop('Unnamed: 0', axis = 1, inplace= True)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

for i in data.columns:
  data[i] = encoder.fit_transform(data[i])
data
```

```python
#THE SUPPORT
#main_support = 0.2, this specifies the minimum support threshold.
#items with support value below 0.2(i.e present in less then 20% of the transaction) will be excluded from the result
#so this shows us the percentage nymber of times occurs in descending order.

support = apriori(data, min_support=0.2, use_colnames=True)
support.sort_values(by = 'support', ascending=False)
```

```python
#THE CONFIDENCE
confidence = association_rules(support, metric='confidence', min_threshold=0.3, support_only= False, num_itemsets= None)
confidence.sort_values(by = 'confidence', ascending=False)
#Zhang's metrics is a measure designed to access the strength of association (positive or negative) bte tow items
#taking into account both their co-occurence and their non co-occurence
#it is particularly useful when we want to understand how the presence or adsence os item A affects the likelyhood of item B being present in a transaction
#in practical terms, Zhang's metrics can be used to access the relationship btw two items or multiple items in transction data
#consider both their co-occurence and their non co-occurence. Positive alues indicate a positive association,
#while negative value suggest a negative association or avoidance. Values close to zero indicate no significant association.
```

#Example 2

```python
df = pd.read_csv('/content/GroceryStoreDataSet.csv', header = None)
df
```

```python
df.rename(columns = {0:'Items'}, inplace= True)
df
```

```python
#TURNING THE ITEM TO A LIST
item_list = df['Items'].to_list()

new_list = []

for i in item_list:
  new_list.append(i.split(','))
new_list
```

```python
#ENCODING THE LIST
from sklearn.preprocessing import LabelEncoder
encoder = TransactionEncoder()

encoded = encoder.fit_transform(new_list)
encoded
```

```python
#TURNING THE ENCODED VALUE INTO A DATAFRAME
#TURN ENCODED DATA TO A DATAFRAME. 'True' to be 1 and 'False' to be 0

encodedData = pd.DataFrame(encoded, columns= encoder.columns_, dtype = int)
encodedData
```

```python
support = apriori(encodedData, min_support=0.2, use_colnames=True)
support.sort_values(by = 'support', ascending=False)
```

```python
#THE CONFIDENCE
confidence = association_rules(support, metric='confidence', min_threshold=0.3, support_only= False, num_itemsets= None)
confidence.sort_values(by = 'confidence', ascending=False)
```
