```python
import pandas as pd

# Load a CSV file
df = pd.read_csv('/content/STEG_BILLING_HISTORY.csv')

# Display the first few rows of the DataFrame
client_0_bills = df.head(10)
print(client_0_bills)
```

```python
#What is the data type of the 'client_0_bills' variable ?
type(client_0_bills)
```

```python
#Display the general information of the dataset and try to answer the following questions
general_info = df.describe(include='all')
print(general_info)
```

```python
#How many rows and columns do we have in this dataset ?
num_rows, num_columns = df.shape

# Display the number of rows and columns
print(f'Number of rows: {num_rows}')
print(f'Number of columns: {num_columns}')
```

```python
# Select columns that are categorical (dtype 'object' or 'category')
categorical_columns = df.select_dtypes(include=['object', 'category']).columns

# Get the number of categorical columns
num_categorical = len(categorical_columns)

# Output the result
print(f"Number of categorical features: {num_categorical}")
print(f"Categorical features: {categorical_columns.tolist()}")
```

```python
# Calculate the memory usage of the dataframe in bytes
memory_usage = df.memory_usage(deep=True).sum()

# Convert memory usage to megabytes (optional)
memory_usage_mb = memory_usage / (1024 ** 2)

# Output the memory usage
print(f"Memory usage: {memory_usage_mb:.2f} MB")
```

```python
# Check for missing values in each column
missing_values = df.isnull().sum()

# Output the columns with their corresponding count of missing values
print(missing_values)
```

I'm using Removing Missing Values strategy because the dataset is large, and the proportion of missing values is small, The rows or columns with missing data are not important for analysis

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Drop columns with missing values
df_cleaned = df.dropna(axis=1)

# Output the cleaned DataFrame
print(df_cleaned)
```

```python
#Method 1: Using Boolean Indexing
client_id = 'train_Client_0'
client_records = df[df['client_id'] == client_id]
print(client_records)
```

```python
#Method 2: Using query Method
client_records_query = df.query("client_id == 'train_Client_0'")
print(client_records_query)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Fit the LabelEncoder to the 'counter_type' column and transform it
df['counter_type_numeric'] = label_encoder.fit_transform(df['counter_type'])

# Output the first few rows to verify the transformation
print(df[['counter_type', 'counter_type_numeric']].head())
```

```python
#Delete the 'counter_statue' feature from the Dataframe
df = df.drop('counter_statue', axis=1)
print(df.head())
```
