```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
```

```python
loan = pd.read_csv('/content/Loan_prediction_dataset.csv')
loan.head()
```

```python
# univariate analysis
# show the gender or educational distribution in the data
# show the distribution of any three numerical variables


# Bivariate Analysis
# what gender has the highest level of graduates
# what educational level makes the most income
# is there a relationship between applicant income and the amount thay intend to borrow?
# what marital status is the most self-employed
```

```python
# show the gender or educational distribution in the data
plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'darkgrid')
sns.countplot(data= loan, x= 'Gender', palette= "tab10")
plt.title('Gender Distribution')
plt.show
```

```python
# show educational distribution in the data
plt.figure(figsize=(5,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'darkgrid')
sns.countplot(data= loan, x= 'Education', palette= "Paired")
plt.title('Education Distribution')
plt.show
```

```python
# show the distribution of any three numerical variables
plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'darkgrid')
sns.kdeplot(data= loan, x= 'ApplicantIncome', palette= "Paired")
plt.title('Applicant Income Distribution')
plt.show
```

```python
# Bivariate Analysis
# what gender has the highest level of graduates

plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'whitegrid')
sns.countplot(data= loan, x= 'Gender', hue = 'Education', palette= "Paired")
plt.title('Gender vs Education')
plt.show
```

```python
# what educational level makes the most income

plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'whitegrid')
fig = sns.barplot(data= loan, x= 'Education', y = 'ApplicantIncome', ci = None, estimator = 'mean', palette= "Paired")
plt.title('Education vs ApplicantIncome')

for i in fig.containers:
    fig.bar_label(i)
plt.show()
```

```python
# is there a relationship between applicant income and the amount thay intend to borrow?

plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'whitegrid')
fig = sns.regplot(data= loan, y= 'LoanAmount', x = 'ApplicantIncome', ci = None)
plt.title(f'{loan.LoanAmount.corr(loan.ApplicantIncome).round(2)}% relationship btw Applicant income and Loan Amount')
plt.show()
```

```python
# what marital status is the most self-employed
plt.figure(figsize=(10,5)) #To adjust the size of the chart. 5(x=width), 5(y=width)
sns.set(style= 'darkgrid')
sns.countplot(data= loan, x= 'Married', hue= 'Self_Employed', palette= "tab10")
plt.title('Marital Status vs Self Employed')
plt.show
```
