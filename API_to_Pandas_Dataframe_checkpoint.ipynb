```python
# Request the Near Earth Object data
# Endpoint for Near Earth Object Web Service (NeoWs)
neo_url = f'https://api.nasa.gov/neo/rest/v1/feed?api_key={api_key}'
```

```python
# Request the data
response = requests.get(neo_url)
neo_data = response.json()
```

```python
neo_data
```

```python
import pandas as pd

# Extract relevant information
asteroids_data = neo_data['near_earth_objects']
asteroid_list = []

for date, asteroids in asteroids_data.items(): # Iterate over dates and corresponding asteroid lists
    for asteroid in asteroids:
        close_approach_data = asteroid['close_approach_data']
        if close_approach_data:
            first_approach = close_approach_data[0]
            asteroid_list.append({
                'Asteroid ID': asteroid['id'],
                'Asteroid name': asteroid['name'],
                'Minimal estimated diameter in Kilometre': asteroid['estimated_diameter']['kilometers']['estimated_diameter_min'],
                'Absolute_magnitude': asteroid['absolute_magnitude_h'],
                'Relative_velocity(km/s)': first_approach['relative_velocity']['kilometers_per_second']
            })

# Create a DataFrame
asteroids_df = pd.DataFrame(asteroid_list)
```

```python
asteroids_df
```

```python
# Data Pre-Processing
# Check for missing values and handle them if necessary
print(asteroids_df.isnull().sum())
```

```python
# Check for duplicate values
asteroids_df.duplicated().sum()
```

```python
# Ensure the data types are appropriate
asteroids_df['Minimal estimated diameter in Kilometre'] = asteroids_df['Minimal estimated diameter in Kilometre'].astype(float)
asteroids_df['Absolute_magnitude'] = asteroids_df['Absolute_magnitude'].astype(float)
asteroids_df['Relative_velocity(km/s)'] = asteroids_df['Relative_velocity(km/s)'].astype(float)
```

```python
# Export to CSV
asteroids_df.to_csv('asteroids_data.csv', index=False)

print("Data has been exported to asteroids_data.csv")
```
