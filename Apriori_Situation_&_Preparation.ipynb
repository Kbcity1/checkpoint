```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apples', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice Cream', 'Eggs']]
```

```python
import mlxtend

import pandas as pd

from mlxtend.preprocessing import TransactionEncoder

te=TransactionEncoder()

te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset

df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array

df
```

```python
#Letâ€™s select itemsets with a minimum of 60% Support

from mlxtend.frequent_patterns import apriori

apriori(df, min_support=0.6)
```

```python
##Instead of column indices we can use column names.
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True)

frequent_itemsets
```

```python
num_itemsets = len(frequent_itemsets)
print(f"Number of frequent itemsets: {num_itemsets}")
```

```python
import mlxtend

import pandas as pd

from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import apriori

from mlxtend.frequent_patterns import association_rules  # Make sure to import it properly

#Confidence Code

#In case we want to extract rules based on other metrics like confidence, we can use association_rules from mlxtend.frequent_patterns library.


rules = association_rules(frequent_itemsets, metric="confidence", num_itemsets=11, min_threshold=0.7) # associate items

print(rules) # Print the generated rules
```

```python
#Associating based on Lift
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="lift", num_itemsets=11, min_threshold=1.25)
```
