```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/startUp (1).csv')
data.head()
```

```python
copy = data.copy()
copy.head()
```

```python
copy.drop(['Unnamed: 0'], axis= 1, inplace= True) #to drop the unnamed column
copy.head()
```

```python
num = copy.select_dtypes(include = 'number') #To show the numerical and categorical data
cat = copy.select_dtypes(exclude = 'number')

print('Numerical Data')
display(num.head())
print('Categorical Data')
display(cat.head())
```

```python
copy2 = copy.drop(['State'], axis = 1)
plt.figure(figsize = (10, 5))
sns.heatmap(copy2.corr(), annot = True, cmap = 'crest')
```

```python
copy2.corr() #to check the correlatio
```

```python
#using scaler to reduce the numerical distance of variables
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = StandardScaler()

for cols in copy.drop(['Profit'], axis = 1):
  if copy[cols].dtype != 'O':
    copy[cols] = scaler.fit_transform(copy[[cols]])

copy.head()
```

```python
#here we encode, turning categorical column into numbers
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

#data['State'] = encoder.fit.transform(data['State'])----- for encoding each lines
#data.head()

#OR


for cols in copy.drop(['Profit'], axis = 1):
  if copy[cols].dtype == 'O':
    copy[cols] = encoder.fit_transform(copy[[cols]])

copy.head()
```

```python
#Modelling
from sklearn.model_selection import train_test_split

x = copy.drop('Profit', axis = 1)
y = copy.Profit

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 90)
print(f'Xtrain shape: {xtrain.shape}')
print(f'Xtest shape: {xtest.shape}')
print(f'Ytrain shape: {ytrain.shape}')
print(f'Ytest shape: {ytest.shape}')
```

```python
#The model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model = LinearRegression()
model.fit(xtrain, ytrain)
model
```

```python
#The prediction
model.predict(xtest)
prediction = model.predict(xtest)
compare = pd.DataFrame()
compare['Actual'] = ytest
compare['Predicted'] = prediction

compare
```

```python
r2_score(ytest, prediction)
```
