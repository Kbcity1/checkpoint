```python
# List to store shopping items
shopping_list = []

# Tuple to store categories of items (immutable)
categories = ("Fruits", "Vegetables", "Dairy", "Bakery", "Meat")

# Set to store unique items bought
items_bought = set()

# Dictionary to store items with additional details (e.g., quantity)
shopping_details = {}

# Function to display the menu
def display_menu():
    print("\nMenu:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View items")
    print("4. View categories")
    print("5. View items bought")
    print("6. View shopping details")
    print("7. Exit")

# Function to add an item to the shopping list
def add_item():
    item = input("Enter the item to add: ")
    quantity = input("Enter the quantity: ")
    shopping_list.append(item)
    items_bought.add(item)
    shopping_details[item] = quantity
    print(f"'{item}' has been added to the shopping list.")

# Function to remove an item from the shopping list
def remove_item():
    item = input("Enter the item to remove: ")
    if item in shopping_list:
        shopping_list.remove(item)
        items_bought.discard(item)
        shopping_details.pop(item, None)
        print(f"'{item}' has been removed from the shopping list.")
    else:
        print(f"'{item}' is not in the shopping list.")

# Function to view items in the shopping list
def view_items():
    if shopping_list:
        print("\nShopping list items:")
        for item in shopping_list:
            print(f"- {item}")
    else:
        print("The shopping list is empty.")

# Function to view categories of items
def view_categories():
    print("\nCategories of items:")
    for category in categories:
        print(f"- {category}")

# Function to view unique items bought
def view_items_bought():
    if items_bought:
        print("\nUnique items bought:")
        for item in items_bought:
            print(f"- {item}")
    else:
        print("No items bought yet.")

# Function to view shopping details
def view_shopping_details():
    if shopping_details:
        print("\nShopping details:")
        for item, quantity in shopping_details.items():
            print(f"- {item}: {quantity}")
    else:
        print("No shopping details available.")

# Main program loop
while True:
    display_menu()
    choice = input("Enter your choice (1-7): ")

    if choice == '1':
        add_item()
    elif choice == '2':
        remove_item()
    elif choice == '3':
        view_items()
    elif choice == '4':
        view_categories()
    elif choice == '5':
        view_items_bought()
    elif choice == '6':
        view_shopping_details()
    elif choice == '7':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 7.")
```
